```python
import pandas as pd
import numpy as np

import neutcurve

from plotnine import *

import csv
import matplotlib.pyplot as plt
```

```python
CBP = neutcurve.colorschemes.CBPALETTE
CBM = neutcurve.colorschemes.CBMARKERS
theme_set(theme_seaborn(style='white', context='paper', font_scale=1))
```

```python
df1 = pd.read_csv('../../../../../../lab_notebooks/kdusenbu/VEP_DMS/SARS_CoV_2_neut_data/AG_full_COVID_003_fracinfect.csv')
df1_15529 = df1[df1['serum'] == '15529'].copy()
df1_15529['virus'] = df1_15529['virus'].replace('Spike-pLV', 'Full Length')
```

```python
df2 = pd.read_csv('../../../../../../lab_notebooks/kdusenbu/VEP_DMS/SARS_CoV_2_neut_data/200604_deltaCT_testneuts.csv')
df2['serum'] = df2['serum'].astype(str)

deltaCT_df = df2[(df2['virus'] == 'delta18') |
                (df2['virus'] == 'delta21') | 
                (df2['virus'] == 'FullLength')].copy()
deltaCT_df['virus'] = deltaCT_df['virus'].replace('FullLength', 'Full_Length_CTtest')
deltaCT_rep1_df = deltaCT_df[deltaCT_df['replicate'] == 1]

FL_prev = df2[df2['virus'] == 'FL_prev'].copy()
FL_prev['virus'] = FL_prev['virus'].replace('FL_prev', 'Full_Length_DKrun')

```

```python
final_df = pd.concat([df1_15529, deltaCT_rep1_df, FL_prev])
final_df['serum'] = final_df['serum'].replace('15529', 'COVID-19 +, 43 days p.s.o.')
display(final_df)
```

```python
final_fits = neutcurve.CurveFits(final_df, fixtop=True)
```

```python
serum = 'COVID-19 +, 43 days p.s.o.'
```

```python
curve = final_fits.getCurve(serum=serum, virus='Full_Length_DKrun', replicate='average')
print(f"The IC50 is {curve.ic50():.3g}")
print(f"This is a dilution of 1:{1/curve.ic50():.4g}")
final_fl2_ic50 = curve.ic50()
fig, ax = curve.plot()
```
